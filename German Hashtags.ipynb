# Let's tidy up this data set!

This is a data set shared with me earlier this year. It is not something I want to pubicly publish anywhere, but I think it is OK to use this list and keep it in a public repo (for a limited time), for practice purposes.

This is a list of hashtags relating to the German Bundestag election in 2021. We used this CSV to create lexica of relevant hashtags, so we do not currently use all the variables that are in this non-tidy data set.

```python
#Import the necessary packages

import numpy as np                   # start by importing the relevant packages/libraries
import pandas as pd                  # that we use throughout this worksheet
import re
import datetime
```

## Multiple variables stored in one column

```python
df_tb=pd.read_csv('German Hashtags.csv')
df_tb.columns

df_tb.sample(5)
```

```python
del df_tb['Date Added']
df_tb
```

```python
df_tb.drop(['Category 2', 'Sub-Category 2', 'Role', 'Real Name', 'Sentiment'], axis = 1)
```

```python
# Applying melt (to a long format)
df_tb=df_tb.melt(id_vars=["Category 1", "Sub-Category 1", "Hashtags"],var_name=["Note category"],value_name="Notes")

df_tb.sample(5)
```

```python
print(df_tb)
```

```python
print(df_tb.Hashtags)
```

## Remove observations which do not have a hashtag associated with them

```python
mod_df_tb = df_tb.dropna( how='all',
                    subset=['Hashtags'])
```
